load the dataset

```python
df = pd.read_csv("linear_regression_train.csv")
df.head()
# print(df.shape)
# print(df.columns)
```

## labeling x and y


```python


df = df.drop(df.index[0])

X = df.drop(columns= ["target"]).values
y = df["target"].values
```

 Train / Validation / Test split (NO randomization)

```python
m = len(X)
train_end = int(0.7 * m)
val_end = int(0.85 * m)

X_train = X[:train_end]
y_train = y[:train_end]

X_val = X[train_end:val_end]
y_val = y[train_end:val_end]

X_test = X[val_end:]
y_test = y[val_end:]




# Normalize y using TRAIN stats only
y_mean = np.mean(y_train)
y_std = np.std(y_train)

y_train = (y_train - y_mean) / y_std
y_val = (y_val - y_mean) / y_std
y_test = (y_test - y_mean) / y_std
```

 regression training

```python
def linear_regression_train(x, y, lr=0.001, epochs = 2000):
    m,n = x.shape
    w = np.zeros(n)
    b = 0

    for i in range(epochs):
        y_pred = np.dot(x,w) + b
        dj_dw = (1/m)*np.dot(x.T,(y_pred - y))
        dj_db =(1/m)* np.sum(y_pred - y)

        w = w - lr*dj_dw
        b = b - lr*dj_db
        if(i%50 == 0):

            print(compute_cost(y_pred,y))
 
    return w,b
```

standardization  featuere

```python
def standardize(x, mean=None, std=None):
    if mean is None:
        mean = np.mean(x, axis=0)
        std = np.std(x, axis=0)
        std[std == 0] = 1
    return (x - mean) / std, mean, std

X_train, x_mean, x_std = standardize(X_train)
X_val, _, _ = standardize(X_val, x_mean, x_std)
X_test, _, _ = standardize(X_test, x_mean, x_std)
```

# train and predict

```python
w, b = linear_regression_train(X_train, y_train)


# predicting
train_preds = np.dot(X_train,w) + b
val_preds = np.dot(X_val, w) + b
test_preds = np.dot(X_test, w) + b
```

```python
def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - ss_res / ss_tot
```

```python
print("Train R2:", r2_score(y_train, train_preds))
print("Validation R2:", r2_score(y_val, val_preds))
print("Test R2:", r2_score(y_test, test_preds))
```
