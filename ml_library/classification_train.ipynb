```python
def logistic_regression_train(x, y, lr = 0.1, epochs= 300):
    m,n = x.shape
    w = np.zeros(n)
    b = 0

    for i in range(epochs):
        z=np.dot(x,w) + b
        y_pred = sigmoid(z)
        dj_dw = (1/m)*np.dot(x.T,(y_pred - y))
        dj_db = (1/m)*np.sum(y_pred - y)

        w = w - lr*dj_dw
        b = b - lr*dj_db
    
    return w,b
```

train data

```python
w_new, b_new = logistic_regression_train(x, y)
```

predicting using trained model

```python
def logistic_predict(x, w, b):
    z = np.dot(x,w) + b
    y_pred = sigmoid(z)
    return (y_pred>=0.5).astype(int)
```

predicting value

```python
y_val_pred = logistic_predict(x,w_new,b_new)
accuracy = np.mean(y_val_pred == y )
print("Validation Accuracy:", accuracy)

print(y)

print(y_val_pred)
```
